Immigration to Canada from Eastern Europe during 1980 - 2013

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
#also needed: openpyxl (install by pip)
```

```python
from dataset import *
```

Analysis

```python
datasetEE = datasetEE[years].head()
datasetEE = datasetEE[years].transpose()
datasetEE.index = datasetEE.index.map(int)
datasetEE
```

Area plot

```python
datasetEE.plot(kind='area', stacked=False, figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

Histogram

```python
import numpy as np
```

```python
data_hist = datasetEE[[1990]].transpose()
data_hist
```

```python
count, bin_edges = np.histogram(data_hist, 8)  #creates 10 bins by default but here 8 is enough
```

```python
data_hist.plot(kind='hist', figsize=(8, 5), bins=8, xticks=bin_edges, stacked=False)

plt.title('Histogram of Immigration to Canada in 1990')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

Bar Chars

```python
df_pol = datasetEE.loc['Poland']
df_pol.head()
```

```python
df_pol.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Polish immigrants to Canada from 1980 to 2013')

plt.annotate('', 
             xy=(10, 17500),  
             xytext=(5, 3000),
             xycoords='data',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

plt.annotate('1985 - 1990 increase',  
             xy=(5, 5000), 
             rotation=72.5,  
             va='bottom',   
             ha='left',
             )

plt.show()
```

```python
df_pol.sort_values(axis=0, ascending=True, inplace=True, kind='quicksort', ignore_index=False,)
df_pol
```

```python
df_pol.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Polish immigrants to Canada from 1980 to 2013')

for index, value in enumerate(df_pol):
  label = format(int(value), ',')
    
  plt.annotate(label, xy=(value - 990, index - 0.2), color='white')

plt.show()
```
