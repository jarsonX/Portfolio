Immigration to Canada from Eastern Europe during 1980 - 2013

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
#also needed: openpyxl (install by pip)
```

```python
from dataset import *
```

Analysis

(1) Pie chart, Immigration to Canada from Eastern Europe

```python
df_for_pie = datasetEE.groupby('Country', axis=0).sum()
df_for_pie
```

```python
colors_list = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']
explode_list = [0.1, 0, 0.1, 0, 0, 0.1, 0, 0, 0.1, 0]

df_for_pie['Total'].plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%', startangle=90, labels=None, pctdistance=1.2, colors=colors_list, explode=explode_list)

plt.title('Immigration to Canada from Eastern Europe for 1980 - 2013', y=1.12)
plt.axis('equal')

plt.legend(labels=df_for_pie.index, loc='upper right')

plt.show()
```

(2) Box plot, Polish and Romanian Immigrants for 1980 - 2013

```python
df_pol_ro = datasetEE.loc[['Poland', 'Romania'], years].transpose()
df_pol_ro
```

```python
df_pol_ro.plot(kind='box', figsize=(8, 6), color='blue')

plt.title('Polish and Romanian Immigrants for 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_pol_ro.describe()
```

Subplots

```python
fig = plt.figure()

ax0 = fig.add_subplot(1, 2, 1)  # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2)

# Subplot 1
df_pol_ro.plot(kind='box', color='blue', vert=False, figsize=(20,6), ax=ax0)  # add to subplot 1
ax0.set_title('Immigrants from Poland and Romania (1980 - 2013), box plot')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2
df_pol_ro.plot(kind='line', figsize=(20,6), ax=ax1)  # add to subplot 2
ax1.set_title('Immigrants from Poland and Romania (1980 - 2013), line plot')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
```
